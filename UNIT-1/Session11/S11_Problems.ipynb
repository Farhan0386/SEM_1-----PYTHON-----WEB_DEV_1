```python
# STRING FORMATTING
# Using format()
a="farhan"
b="mohan"
print("{} and {} are good friends.".format(a,b))
# Using f-strings
c=6.32456
print(f"Height of {a} and {b} is {c:.2f}")  # c:.2f means round of the height by 2 decimal
print(f"Height of {a} and {b} is {c:.4f}") 
```

```python
a="Farhan Hussain , Saddam Hussain"
b="$$25.00$$"
print(a.upper())
print(a.lower())
print(a.capitalize())
print(b.strip('$$'))  #it remove charaters that i want from string from beging and starting
print(a.replace("Farhan","Saddam")) #it does not change the orignal string 
print(a)
print(a.split(","))
print(a.find("Hussain")) # it provide index of Hussain in string at index 7 hussain starts so it return 7 there are two it return of first Hussain
print(a.index("Hussain")) # it give index
print(a.count("s"))
print(a.startswith("F"))
print(a.endswith("n"))
print(a.isalpha())  #Returns True if the string has at least one character and every character is alphabetic (letters). Otherwise returns False.
print(a.isalnum()) #Returns True if the string has at least one character and every character is alphanumeric (letters or digits). Otherwise returns False.
```
