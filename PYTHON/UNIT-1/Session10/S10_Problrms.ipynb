```python
# logical operators 
a=True
b=True
c=False
d=False
print("a and b :" , a and b)
print("a and c :" , a and c)
print("c and d :" , c and d)
print("a or b :" , a or b)
print("a or c :" , a or c)
print("c or d :" , c or d)
```

```python
#identity operators
a=3
b=3
c=4
print("a is b :" , a is b)
print("a is c :" , a is c)
print("b is c :" , b is c)
print("a is not c :" , a is not c)
```

```python
#membership operators
a=[1,2,3,4,5,6]
b="Farhan"
c=(1,2,3,4,5,6)
d={"name":"farhan","gender":"male","country":"INDIA"}
print("4 i a :", 4 in a)
print("7 in a:",7 in a)
print("h in b :",'h' in b)
print("4 in c:",4 in c)
print("name in d :",'name' in d)
```

```python
#bitwise operators
a=5  # 101
b=10  # 1010
print("a & b :",a & b)
print("a | b :",a | b)
print(" ~ b  :", ~ b)
print(" ~ a  :", ~ a)
print("a ^ b :",a ^ b)
print("a << 1 :",a<<1)
print("a << 2 :",a<<2)
print("a >> 1 :",a>>1)
print("a >> 2 :",a>>2)
```
